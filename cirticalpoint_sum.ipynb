```python
import openpyxl #패키지 불러오기
from openpyxl import Workbook
import re
import pandas as pd
filename = "only_criticalpoint.xlsx" #파일명
book = openpyxl.load_workbook(filename) #엑셀파일 book 변수에 저장
```

```python
#sheet 추출
sheet=book.worksheets[0] #첫번쨰 워크시트 저장
```

```python
#데이터 입력하기
data = [] #리스트 자료형 생성
for row in sheet.rows: #전체 행에 대하여 반복실행
    data.append([
        #row[1].value, #PM10 > 80 
        #row[2].value, #PM2.5 > 50 
        row[3].value, #PM2.5 > 80
     ])
```

```python
data
```

```python
critic = data[1:] #data의 첫값(열의 제목) 제외
```

```python
sum = [[0 for col in range(1)] for row in range(269)]
for i in range(len(critic)//7):
   #sum_element = 0
   for j in range(7):
        sum[i] += critic[i*7 + j]
```

```python
import random
number = []
df = pd.DataFrame(columns=['mon', 'tue', 'wedn', 'thurs', 'fri', 'satur', 'sun'])
for i in range(len(sum)):
       mon = sum[i][0]
       tue = sum[i][1]
       wedn = sum[i][2]
       thurs = sum[i][3]
       fri = sum[i][4]
       satur = sum[i][5]
       sun = sum[i][6]
       # DataFrame에 특정 정보를 이용하여 data 채우기
       df = df.append(pd.DataFrame([[mon,tue,wedn,thurs,fri,satur,sun]], columns=['mon', 'tue', 'wedn', 'thurs', 'fri', 'satur', 'sun']), ignore_index=False)
print(df)   
```
