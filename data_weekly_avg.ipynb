```python
import openpyxl #패키지 불러오기
from openpyxl import Workbook
import re
import pandas as pd
filename = "finedust_daily_elements_from2015.xlsx" #파일명
book = openpyxl.load_workbook(filename) #엑셀파일 book 변수에 저장
```

```python
#sheet 추출
sheet=book.worksheets[0] #첫번쨰 워크시트 저장
```

```python
#데이터 입력하기
data = [] #리스트 자료형 생성
for row in sheet.rows: #전체 행에 대하여 반복실행
    data.append([
         row[0].value, #년도
         row[1].value, #월
         row[2].value, #일
         row[3].value, #입자명 데이터
         row[29].value, #입자 개수 평균 데이터
         #row[32].value, #날짜 데이터 -> 엑셀 내 함수 CONCATENATE로 날짜를 구하여서 python에서 값 확인불가 
     ])
```

```python
data
```

```python
#주간 평균값 
weekly_avg = [] 
for row in sheet.rows:
    weekly_avg.append([
        row[30].value
    ])
```

```python
weekly_avg
```

```python
avg = []
for i in range(1, len(weekly_avg)):
    if (i%7)==0:
        k = i-2
        avg += weekly_avg[k]        
```

```python
avg
```

```python
import random
# 빈 DataFrame 생성하기
df = pd.DataFrame(columns=['index', 'number'])
for index in range(1, 1348):
       number = avg[index]
       # DataFrame에 특정 정보를 이용하여 data 채우기
       df = df.append(pd.DataFrame([[index, number]], columns=['idx', 'number']), ignore_index=False)
df.set_index('index', inplace=True)
print(df)   
```
